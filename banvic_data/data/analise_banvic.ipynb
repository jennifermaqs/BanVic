3 - Análise por dia da semana

```python
df_completo['ano'] = df_completo['data_transacao'].dt.year
df_completo['mes'] = df_completo['data_transacao'].dt.month
df_completo['dia_da_semana_en'] = df_completo['data_transacao'].dt.day_name()

mapa_meses = {
    1: 'Janeiro', 2: 'Fevereiro', 3: 'Março', 4: 'Abril',
    5: 'Maio', 6: 'Junho', 7: 'Julho', 8: 'Agosto',
    9: 'Setembro', 10: 'Outubro', 11: 'Novembro', 12: 'Dezembro'
}

mapa_dias = {
    'Monday': 'Segunda-feira', 'Tuesday': 'Terça-feira', 'Wednesday': 'Quarta-feira',
    'Thursday': 'Quinta-feira', 'Friday': 'Sexta-feira', 'Saturday': 'Sábado',
    'Sunday': 'Domingo'
}

df_completo['nome_mes'] = df_completo['mes'].map(mapa_meses)
df_completo['dia_da_semana'] = df_completo['dia_da_semana_en'].map(mapa_dias)


df_completo[['data_transacao', 'ano', 'nome_mes', 'dia_da_semana']].head()
```

```python
dias_unicos = df_completo['dia_da_semana'].unique()

resultados = []

for dia_especifico in dias_unicos:

    transacoes_do_dia = df_completo[df_completo['dia_da_semana'] == dia_especifico]

    total_transacoes_dia = len(transacoes_do_dia)

    volume_total_dia = transacoes_do_dia['valor_transacao'].sum()

    resultados.append([dia_especifico, total_transacoes_dia, volume_total_dia])


analise_semana = pd.DataFrame(
    resultados,
    columns=['dia_da_semana', 'total_transacoes', 'volume_total_movimentado']
)


print(analise_semana)
```

3.1. Qual o dia da semana em que a média de transações aprovadas é maior e qual tem, também em média, maior volume movimentado?


```python
transacoes_volume = analise_semana.sort_values(by='total_transacoes', ascending=False)
print("\n Número de Transações e volume total movimentado:")
print(transacoes_volume)

print("\n")
```

3.2. Meses que são pares o volume médio de transações é significativamente maior do que os meses ímpares?

```python
df_completo['paridade'] = df_completo['mes'].apply(lambda x: 'Par' if x % 2 == 0 else 'Ímpar')

#verificacao
print(df_completo[['nome_mes', 'paridade']].head())
```

```python
#transações em meses pares
transacoes_mes_par = df_completo[df_completo['paridade'] == 'Par']

#transações em meses impares
transacoes_mes_impar = df_completo[df_completo['paridade'] == 'Ímpar']


#media do valor da transacao de cada 
media_mes_par = transacoes_mes_par['valor_transacao'].mean()
media_mes_impar = transacoes_mes_impar['valor_transacao'].mean()


#valor medio da transacao por paridade
print(f"pares: R$ {media_mes_par:.2f}")
print(f"impares: R$ {media_mes_impar:.2f}")

print("\n")
```

4. Taxa de câmbio. Existe alguma relação do valor da moeda com o volume e o valor das transações (BCB)

```python
from bcb import sgs
from datetime import datetime
```

```python
data_inicio = df_completo['data_transacao'].min()
data_fim = df_completo['data_transacao'].max()

print(f"{data_inicio.date()} - {data_fim.date()}")

lista_dfs_dolar = []

for ano in range(data_inicio.year, data_fim.year + 1, 8):
    start_chunk = datetime(ano, 1, 1)
    end_chunk = datetime(min(ano + 7, data_fim.year), 12, 31)


    df_chunk = sgs.get({'dolar': 1}, start=start_chunk, end=end_chunk)

    lista_dfs_dolar.append(df_chunk)

df_dolar_bcb = pd.concat(lista_dfs_dolar)

df_dolar_bcb = df_dolar_bcb.reset_index()
df_dolar_bcb = df_dolar_bcb.rename(columns={'Date': 'data', 'dolar': 'valor_dolar'})
```

```python
print(df_dolar_bcb.head())
```

```python
print(df_dolar_bcb.tail())
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tabela_de_busca_dolar = df_dolar_bcb.set_index('data')['valor_dolar']


df_completo['data_para_busca'] = pd.to_datetime(df_completo['data_transacao'].dt.date)


df_completo['valor_dolar'] = df_completo['data_para_busca'].map(tabela_de_busca_dolar)


df_final = df_completo.dropna(subset=['valor_dolar']).copy()

correlacao = df_final['valor_dolar'].corr(df_final['valor_transacao'])
print(f"\ncoeficiente de correlação: {correlacao:.4f}")

print("Gráfico de dispersão")
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df_final, x='valor_dolar', y='valor_transacao', alpha=0.1, color='gray')
plt.title('Relação entre Cotação do Dólar e Valor da Transação')
plt.xlabel('Valor do Dólar (R$)')
plt.ylabel('Valor da Transação (R$)')
plt.savefig('grafico_dolar.png', dpi=300)
plt.show()



print("\n")
```
